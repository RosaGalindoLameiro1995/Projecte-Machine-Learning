## Transformació de les dades amb PUT()
A continuació intentarem actualitzar les dades del servidor utilitzant el mètode put().

```python
URL = "https://www.swapi.tech/api/planets/1"
```

```python
DATA = { 
    "diameter": 10000,
    "terrain": "Seco",
    "name": "Inventado"
}

HEADERS = {"Authorization": "Bearer TU_TOKEN"}
respuesta = requests.put(URL, json=DATA, headers=HEADERS)

if respuesta.status_code == 200:
    data = respuesta.json()
    print('Datos actualizados')
    print('Respuesta', data)
else:
    print('Error al actualizar los datos', respuesta.text)
    
```

## Web Scrapping de Páginas Amarillas.

```python
import requests
from bs4 import BeautifulSoup
```

```python
URLB = "https://www.paginasamarillas.com.co/servicios/restaurantes"
pag = requests.get(URLB)

soup = BeautifulSoup(pag.content, "html.parser")
```

Importem la biblioteca Beautiful Soup, que ens permet gestionar i cercar els resultats obtinguts de manera més fàci.

print(pag.text)

A continuació buscarem algun element utilitzant la biblioteca importada i cercant per l'id de l'element de la pàgina HTML.

```python
resultat = soup.find(id="__NEXT_DATA__")
print(resultat.prettify())
```
